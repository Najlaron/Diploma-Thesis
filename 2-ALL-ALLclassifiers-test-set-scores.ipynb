# We train ALL models on the WHOLE train-set and test them on test-set

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn import neighbors
from ast import literal_eval
```

```python
train_data = pd.read_csv('train_data-CLEANED-7.csv', sep = ';', converters={"embeddings_uncsd-BERT": literal_eval, "embeddings_csd-BERT": literal_eval, "embeddings_USE": literal_eval, "embeddings_MiniLM-6": literal_eval, "embeddings_MiniLM-12": literal_eval, "embeddings_Doc2Vec-enwiki": literal_eval, "embeddings_Doc2Vec-apnews": literal_eval, "embeddings_IS-GloVe": literal_eval, "embeddings_IS-FastText": literal_eval, "embeddings_uncsd-BERT-1024": literal_eval}, dtype={'affil_index': 'int32'})
train_data.drop(columns = ['Unnamed: 0'], inplace = True)
train_data
```

```python
test_data = pd.read_csv('test_data-CLEANED-7.csv', sep = ';', converters={"embeddings_uncsd-BERT": literal_eval, "embeddings_csd-BERT": literal_eval, "embeddings_USE": literal_eval, "embeddings_MiniLM-6": literal_eval, "embeddings_MiniLM-12": literal_eval, "embeddings_Doc2Vec-enwiki": literal_eval, "embeddings_Doc2Vec-apnews": literal_eval, "embeddings_IS-GloVe": literal_eval, "embeddings_IS-FastText": literal_eval, "embeddings_uncsd-BERT-1024": literal_eval}, dtype={'affil_index': 'int32'})
test_data.drop(columns = ['Unnamed: 0'], inplace = True)
test_data
```

```python
X_list_names = train_data.columns[3:].to_list()
print(X_list_names)
```

```python
X_list = []
for name in X_list_names:
    lst = train_data.loc[:, name].to_list()
    X_list.append(lst)
X_list_test = []
for name in X_list_names:
    lst = test_data.loc[:, name].to_list()
    X_list_test.append(lst)
```

```python
y = train_data.loc[:, "labels"].tolist()
y_test = test_data.loc[:, "labels"].tolist()
```

### Training + Testing

```python
results = []
for i, embeddings in enumerate(X_list):
    scores = []
    for rs in [1, 2, 3, 4, 42]:
        clf = MLPClassifier(random_state=rs, max_iter=300).fit(embeddings, y)
        scores.append(clf.score(X_list_test[i], y_test))
        clf = RandomForestClassifier(random_state=rs, n_estimators = 1000, max_depth = 20).fit(embeddings, y)
        scores.append(clf.score(X_list_test[i], y_test))
        clf = make_pipeline(StandardScaler(), SVC(random_state = rs, gamma='auto')).fit(embeddings, y)
        scores.append(clf.score(X_list_test[i], y_test))
        n_neighbors  = 3 #performed best (when testing 1,3,5 and rest (7+) was significantly worse)
        clf = neighbors.KNeighborsClassifier(n_neighbors).fit(embeddings, y)
        scores.append(clf.score(X_list_test[i], y_test))
    results.append(scores)
    print(X_list_names[i], "done")
    print(scores)
results  
#took slightly over 1.5 hour to run
```

```python
ress = []
for res in results:
    nn = []
    rf = []
    svm = []
    knn = []
    for i in range(0, len(res), 8):
        print(res)
        nn.append(res[i])
        rf.append(res[i+1])
        svm.append(res[i+2])
        knn.append(res[i+4][0])
    ress.append([nn, rf, svm, knn])
#print(len(ress[0]))
res_df = pd.DataFrame(ress, index = X_list_names, columns = ['nn', 'rf', 'svm', 'knn'])
```
