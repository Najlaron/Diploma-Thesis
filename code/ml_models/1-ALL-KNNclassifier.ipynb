# All models with K Nearest Neighbours Classifier

```python
import numpy as np
import pandas as pd
#import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import neighbors
from sklearn.model_selection import cross_val_score
from ast import literal_eval
```

```python
train_data = pd.read_csv('train_data-CLEANED-7.csv', sep = ';', converters={"embeddings_uncsd-BERT": literal_eval, "embeddings_csd-BERT": literal_eval, "embeddings_USE": literal_eval, "embeddings_MiniLM-6": literal_eval, "embeddings_MiniLM-12": literal_eval, "embeddings_Doc2Vec-enwiki": literal_eval, "embeddings_Doc2Vec-apnews": literal_eval, "embeddings_IS-GloVe": literal_eval, "embeddings_IS-FastText": literal_eval, "embeddings_uncsd-BERT-1024": literal_eval}, dtype={'affil_index': 'int32'})
train_data.drop(columns = ['Unnamed: 0'], inplace = True)
train_data
```

```python
# test_data = pd.read_csv('test_data-CLEANED-7.csv', sep = ';', converters={"embeddings_uncsd-BERT": literal_eval, "embeddings_csd-BERT": literal_eval, "embeddings_USE": literal_eval, "embeddings_MiniLM-6": literal_eval, "embeddings_MiniLM-12": literal_eval, "embeddings_Doc2Vec-enwiki": literal_eval, "embeddings_Doc2Vec-apnews": literal_eval, "embeddings_IS-GloVe": literal_eval, "embeddings_IS-FastText": literal_eval, "embeddings_uncsd-BERT-1024": literal_eval}, dtype={'affil_index': 'int32'})
# test_data.drop(columns = ['Unnamed: 0'], inplace = True)
# test_data
```

### Different embeddings

```python
X_list_names = train_data.columns[3:].to_list()
print(X_list_names)
```

```python
X_list = []
for name in X_list_names:
    lst = train_data.loc[:, name].to_list()
    X_list.append(lst)
# X_list_test = []
# for name in X_list_names:
#     lst = test_data.loc[:, name].to_list()
#     X_list_test.append(lst)
```

#### Labels

```python
y = train_data.loc[:, "labels"].tolist()
#y_test = test_data.loc[:, "labels"].tolist()
```

### Calculations for all pretrained embeddings models

```python
n_neighbors = 3 #score the best out of [1, 3, 5, 7, 9] overall
```

```python
results = []
#results_test = []
for i, embeddings in enumerate(X_list):
    scores = []
    #test_scores = []
    for rs in [1, 2, 3, 4, 42]: #does not depend on random_state, no need to calculate 5 times
        clf = neighbors.KNeighborsClassifier(n_neighbors) #.fit(embeddings, y) #unnecessary, happens in cross_val_score
        scs = cross_val_score(clf, embeddings, y, cv=5) 
        scores.extend(scs)
        #calculate scores for model using test_data
        #test_scores.append(clf.score(X_list_test[i], y_test))
        #results_test.append([X_list_names[i], test_scores])
    name = X_list_names[i]
    mean = np.mean(scores)
    std = np.std(scores)
    mxm = np.max(scores)
    #avg_test = np.mean(test_scores)
    #results.append([name, scores, mean, std, mxm, test_scores, avg_test])
    results.append([name, scores, mean, std, mxm])
    #print("For %s: %0.2f accuracy with a standard deviation of %0.2f with maximum of: %0.2f and scoring: %0.2f on test-set." % (name, mean, std, mxm, avg_test))
    print("For %s: %0.2f accuracy with a standard deviation of %0.2f with maximum of: %0.2f"% (name, mean, std, mxm))

    #when un/commented this would also test every model on the test-set (only 5 scores)
    #took only 1.5 minutes to run which is many times faster than training other models
```

```python
#results_df = pd.DataFrame(results, columns = ['name', 'scores', 'mean', 'std', 'max', 'test_scores', 'avg_test'])
results_df = pd.DataFrame(results, columns = ['name', 'scores', 'mean', 'std', 'max'])
results_df.to_csv('1-ALL-KNNclassifier-CLEANED-AVG.csv', sep = ';')
results_df
```
