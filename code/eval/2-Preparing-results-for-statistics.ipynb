# Combining results from different models into one data-set of results

```python
import pandas as pd
import numpy as np
from ast import literal_eval
```

```python
data1 = pd.read_csv('1-ALL-NNclassifier-CLEANED-AVG.csv', sep =';')
data1 = data1.drop(['Unnamed: 0'], axis=1)
data1
```

```python
data2 = pd.read_csv('1-ALL-RFclassifier-CLEANED-AVG.csv', sep =';')
data2 = data2.drop(['Unnamed: 0'], axis=1)
data2
```

```python
data3 = pd.read_csv('1-ALL-SVclassifier-CLEANED-AVG.csv', sep =';')
data3 = data3.drop(['Unnamed: 0'], axis=1)
data3
```

```python
data4 = pd.read_csv('1-ALL-KNNclassifier-CLEANED-AVG.csv', sep =';')
data4 = data4.drop(['Unnamed: 0'], axis=1)
data4
```

```python
# append all scores from different classifiers for the same model (name) into one list of scores
data1['scores'] = data1['scores'].apply(literal_eval)
data2['scores'] = data2['scores'].apply(literal_eval)
data3['scores'] = data3['scores'].apply(literal_eval)
data4['scores'] = data4['scores'].apply(literal_eval)
data = pd.DataFrame()
data['name'] = data1['name']
data['scores'] = data1['scores'] + data2['scores'] + data3['scores'] + data4['scores']
data

```

```python
data['std'] = data['scores'].apply(np.std)
data['mean'] = data['scores'].apply(np.mean)
data['median'] = data['scores'].apply(np.median)
data['min'] = data['scores'].apply(np.min)
data['max'] = data['scores'].apply(np.max)
```

```python
data
```

```python
data.to_csv('2-ALL-ALLclassifiers-results.csv', sep=';')
```

```python
#for every score in scores list add a new column into the dataframe
df = pd.DataFrame()
df['name'] = data['name']
column_names = ['nn', 'rf', 'sv', 'knn']
ln_model = int(len(data['scores'][0])/len(column_names)) #100/4 =25
nom = 5 #number of measurements (random seeds)
kcv = 5 #number of folds in k-fold cross validation
#name columns 0nn0, 1nn0, ... , 4nn0, 0nn1, ... , 4nn4, 0rf0, ...
names = []
for c_nm in column_names:
    for i in range(kcv):
        for j in range(nom):
            names.append(str(j) + c_nm + str(i))
# print(len(names))
# print(names)
scores = []
for i in range(len(data['scores'])):
    for j in range(len(data['scores'][i])):
        scores.append(data['scores'][i][j])
# print(len(scores))
# print(scores)
df[names] = np.array(scores).reshape(len(data['scores']), len(names))
df
```

```python
df.to_csv('2-ALL-ALLclassifiers-results-scores.csv', sep=';')
```

# Preparing data for First ANOVA and box-plots (PWE models)

```python
#swap columns with rows
dft = df.set_index('name').T
dft
```

```python
dft.to_csv('2-ALL-ALLclassifiers-results-scores-TRANSPOSED.csv', sep=';')
```

```python
#two columns: dependent variable (score) and classification by PWE model (columns in dft)
df_anova = pd.DataFrame()
columns = dft.columns.tolist()
print(columns)

for column in columns:
    for row in dft.index:
        df_anova = df_anova.append({'name': row, 'score': dft[column][row], 'class': column}, ignore_index=True)
df_anova
```

```python
df_anova.to_csv('2-ALL-ALLclassifiers-results-scores-ANOVA.csv', sep=';')
```

# Preparing data for Second ANOVA and box-plots (ML Classifiers)

```python
candidate_models = ['embeddings_csd-BERT', 'embeddings_uncsd-BERT', 'embeddings_IS-GloVe', 'embeddings_USE', 'embeddings_MiniLM-6', 'embeddings_MiniLM-12'] #models, that scored highest and have no statistically significant difference amongst eachother
df_anova = df_anova[df_anova['class'].isin(candidate_models)]
df_anova
```
