# all-MiniLM-L6-v2

### Models available from SBERT.net

```python
from sentence_transformers import SentenceTransformer
import pandas as pd
import numpy as np
from ast import literal_eval

input_file = 'test_data-CLEANED-5.csv' #'train_data-CLEANED-5.csv'
output_file = 'test_data-CLEANED-6.csv' #'train_data-CLEANED-6.csv'
```

```python
#https://www.sbert.net/docs/pretrained_models.html
#https://github.com/UKPLab/sentence-transformers/blob/master/docs/pretrained_models.md
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
query_embedding = model.encode('How big is London')
passage_embedding = model.encode(['London has 9,787,426 inhabitants at the 2011 census',
                                  'London is known for its finacial district'])
```

```python
#print(query_embedding)
print(query_embedding.shape)
#print(passage_embedding)
print(passage_embedding.shape)
```

```python
t_data = pd.read_csv(input_file, sep = ';')
t_data.drop(columns = ['Unnamed: 0'], inplace = True)
t_data
```

```python
parts = t_data.iloc[:, 0].tolist()
parts
```

```python
try:
    embeddings = model.encode(parts).tolist()
except:
    print("error")
t_data["embeddings_MiniLM-6"] = embeddings
```

```python
t_data
```

```python
t_data.to_csv(output_file, sep = ';')
```
